It is surprising that even in the idealized case where the labels are generated by $y= f(W^\ast,x_i)_+$, and features are i.i.d.Gaussian distributed i.e.,$$x_i \sim \mathcal{N}(0,\mathbb{I}_d),i=1,...,n $$ The  relaxation  fails to recover the correct weight vector $W^\ast$ .

It is possible to recover $W^\ast$ using a very simple procedure  as  long  as n > 2d ,  and n is  large  enough.  We can consider the subset of labels which are strictly positive,$$S=\bigcup_{i \in \{1,...,\},y_i > 0 }\{i\}$$ And attempt to solve for $W^\ast$ using the  pseudoinverse  via $$\hat{W}=X_S^\dagger Y_S$$ Itâ€™s  straightforward  to show that $W^\ast=\hat{W}$ as long as $\mid S \mid \geq d $ and $X_s$ as full column rank.  For i.i.d.  Gaussian features, this holds with high probability when n is large enough.


```
import numpy as np
w=np.random.normal(loc=0.0, scale=1.0, size=(50))
X=np.random.normal(loc=0.0, scale=1.0, size=(100,50))
Y=X@w
```

```
idx=[i for i,u in enumerate(Y)  if u >0]
```

```
w_hat=np.linalg.pinv(X[idx,:])@Y[idx]
```

```
np.sum((w_hat-w)**2)
```
